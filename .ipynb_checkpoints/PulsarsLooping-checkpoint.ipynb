```python
# Write your mean_fits function here:
from astropy.io import fits
import numpy as np
```

```python
def mean_fits(files):
    n = len(files)
    print(n)
    if n > 0:
        hdulist = fits.open(files[0])
        print(hdulist)
        data = hdulist[0].data
        print(data)
        hdulist.close()
    
        for i in range(1, n):
            hdulist = fits.open(files[i])
            print(hdulist)
            data += hdulist[0].data
            hdulist.close()
            
        mean = data / n
        return mean
```
